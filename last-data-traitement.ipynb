```python
import pandas as pd
import re
```

```python
path = r'C:\Users\PC\Downloads\Data_GenerateQuestion\qg_valid.csv'
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
def remove_answer_tag(text):
    cleaned_text = re.sub(r"<answer>.*?(<context>|$)", r"\1", text, flags=re.DOTALL)
    return cleaned_text.strip()

df['text'] = df['text'].apply(remove_answer_tag)

df.to_csv('cleaned_data_type.csv', index=False)
```

```python
df.rename(columns={'text': 'context'}, inplace=True)
```
