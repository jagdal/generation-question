```python
import string
```

```python
def clean_text(text):
    # Supprimer les balises <answer> et <context>
    text = text.replace('<answer>', '').replace('<context>', '')
    # Convertir en minuscules
    text = text.lower()
    # Supprimer la ponctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
```

```python
# Supprimer les stopwords
stop_words = set(stopwords.words('english'))
data['tokens_cleaned'] = data['tokens'].apply(lambda tokens: [word for word in tokens if word not in stop_words])
```

```python
# Lemmatisation avec spaCy
def lemmatize(tokens):
    doc = nlp(" ".join(tokens))
    return [token.lemma_ for token in doc]
```

```python
data['lemmas'] = data['tokens_cleaned'].apply(lemmatize)
```
